```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
```

```python
x_data_generated, y_data_generated = make_classification(scale=1)
```

```python
pd.DataFrame(x_data_generated)
```

```python
 pd.DataFrame(y_data_generated)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
way_ch_features1 = 'Все признаки'
n_features1 = 20
score1 = cross_val_score(LogisticRegression(), x_data_generated, y_data_generated, scoring='accuracy').mean()
score1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import VarianceThreshold
```

```python
plt.figure(figsize=(16, 9))
sns.heatmap(pd.DataFrame(x_data_generated).corr(), annot=True)
```

```python
x = pd.DataFrame(x_data_generated)[[3, 4, 7, 8, 9, 10, 13, 15]]
y = y_data_generated
```

```python
way_ch_features2a = 'Матрица корреляции'
n_features2a = 8
score2a = cross_val_score(LogisticRegression(), x, y, scoring='accuracy').mean()
score2a
```

```python
x = pd.DataFrame(VarianceThreshold(threshold=1.1).fit_transform(x_data_generated))
x
```

```python
way_ch_features2b = 'VarianceThreshold'
n_features2b = 7
score2b = cross_val_score(LogisticRegression(), x, y, scoring='accuracy').mean()
score2b
```

```python
from sklearn.feature_selection import SelectKBest, chi2, f_classif
```

```python
x = pd.DataFrame(SelectKBest(f_classif, k=5).fit_transform(x_data_generated, y_data_generated))
x
```

```python
way_ch_features3 = 'Дисперсионный анализ'
n_features3 = 5
score3 = cross_val_score(LogisticRegression(), x, y, scoring='accuracy').mean()
score3
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Lasso
```

```python
x = pd.DataFrame(SelectFromModel(LogisticRegression()).fit_transform(x_data_generated, y_data_generated))
x
```

```python
way_ch_features4a = 'Логистическая регрессия'
n_features4a = 6
score4a = cross_val_score(LogisticRegression(), x, y, scoring='accuracy').mean()
score4a
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(pd.DataFrame(x_data_generated), y_data_generated)
feat_importances = pd.Series(model.feature_importances_, index=pd.DataFrame(x_data_generated).columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
x = pd.DataFrame(x_data_generated)[[2, 3, 4, 7, 11, 19]]
x
```

```python
way_ch_features4b = 'Случайный лес'
n_features4b = 6
score4b = cross_val_score(LogisticRegression(), x, y, scoring='accuracy').mean()
score4b
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
sfs = SequentialFeatureSelector(knn, n_features_to_select=7)
sfs.fit(x_data_generated, y_data_generated)
x = pd.DataFrame(sfs.transform(x_data_generated))
x
```

```python
way_ch_features5 = 'Перебор признаков'
n_features5 = 7
score5 = cross_val_score(LogisticRegression(), x, y, scoring='accuracy').mean()
score5
```

```python
print(way_ch_features1, n_features1, score1, sep='|')
print(way_ch_features2a, n_features2a, score2a, sep='|')
print(way_ch_features2b, n_features2b, score2b, sep='|')
print(way_ch_features3, n_features3, score3, sep='|')
print(way_ch_features4a, n_features4a, score4a, sep='|')
print(way_ch_features4b, n_features4b, score4b, sep='|')
print(way_ch_features5, n_features5, score5, sep='|')
```

Независимо от способа отбора признаков, результат будет лучше, чем если брать все признаки. Лучше всего показали сбея методы с лолгистической регрессией и с VarianceThreshold. Хуже всего выбор вручную с помощью матрицы корреляции. Возможно это связано с трудностью восприятия матрицы 20*20.
