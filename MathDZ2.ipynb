```python
import numpy as np
```

```python
A1 = np.random.rand(5, 5)
A1
```

```python
print(np.linalg.matrix_rank(A1))
print(np.linalg.det(A1))
print(np.linalg.eig(A1))
```

```python
T = A1.T
T
```

```python
print(np.linalg.matrix_rank(T))
print(np.linalg.det(T))
print(np.linalg.eig(T))
```

При транспонировании ранг, определитель и собственные значения матрицы не меняются.

```python
A1I = np.linalg.inv(A1)
A1I
```

```python
print(np.linalg.matrix_rank(A1I))
print(np.linalg.det(A1I))
print(np.linalg.eig(A1I))
```

Обратная матрица имеет тот же ранг, что и исходная.

```python
I = np.eye(4, 4)
I
```

```python
print(np.linalg.matrix_rank(I))
print(np.linalg.det(I))
print(np.linalg.eig(I))
```

Ранг единичной матрицы зависит от её размерности. Определитель единичной матрицы всегда равен 1. Собсвенные значения единицы.

```python
A2 = np.random.rand(4, 4)
A2
```

```python
print(np.linalg.matrix_rank(A2))
print(np.linalg.det(A2))
print(np.linalg.eig(A2))
```

```python
DA = A2 * I
DA
```

```python
print(np.linalg.matrix_rank(DA))
print(np.linalg.det(DA))
print(np.linalg.eig(DA))
```

Ранг диагональной матрицы зависти от её размерности(или числа элементов в её диагонали). Как и единичной матрицей это проистекает, что вектора n размерности, где координаты каждого вектора равны 0, кромей одной, и их индексы(местоположения) не совпадают. В этом случае они будут линейно независимы, а число линейно независимых векторов явялется рангом матрицы. Пример подобной линейной комбинации векторов - ортогональный базис для 3-х мерного пространства(i, j, k)

Собственные векторы диагональной матрице равны единичной матрице.

```python
O = np.zeros([3, 3])
O
```

```python
print(np.linalg.matrix_rank(O))
print(np.linalg.det(O))
print(np.linalg.eig(O))
```

Ранг и определитель и собственные значения нулевой матрицы равны 0, а собственные векторы единичной матрице.
