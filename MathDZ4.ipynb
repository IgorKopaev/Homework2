```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-5, 5, 0.01)
```

```python
y=np.sin(2*x+1)**5
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Функция')
plt.show()
```

```python
y1 = 10*(np.sin(2*x+1)**4)*np.cos(2*x+1)
```

```python
plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Первая производная')
plt.show()
```

```python
y2 = 80*(np.sin(2*x+1)**3)*(np.cos(2*x+1)**2)-20*(np.sin(2*x+1)**5)
```

```python
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Вторая производная')
plt.show()
```
