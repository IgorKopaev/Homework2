```python
def rastrigin(x):
    return 10.0 * len(x) + np.sum(x**2.0-10.0*np.cos(2.0*np.pi*x), axis=0)
```

```python
n = len(np.arange(-5.12, 5.12, 0.01))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

fig = plt.figure(figsize=[15, 12])
ax = fig.gca(projection='3d')
ax.view_init(45, 30)

X = np.arange(-5.12, 5.12, 0.01)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
Z = rastrigin(np.array([X,Y]))

surf = ax.plot_surface(X, Y, Z, cmap='bwr')
plt.show()
```

```python
from scipy.optimize import minimize
```

```python
x0 = x0=np.random.uniform(low=-5,high=5,size=(2,))
```

```python
x0
```

```python
res = minimize(rastrigin, x0, method='nelder-mead',options={'disp':True})
```

```python
res
```

```python
it_list = []
res_list =[]
for i in range(100):
    x0 = np.random.uniform(low=-5,high=5,size=(2,))
    res = minimize(rastrigin, x0, method='nelder-mead',options={'disp':False})
    it_list.append(res['nit'])
    res_list.append(res['fun'])
print(np.mean(res_list))
print(min(res_list))
print(np.mean(it_list)) 
```
