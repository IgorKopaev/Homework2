```python
logs = pd.read_csv('visit_log2.csv')
```

```python
logs.head()
```

```python
def type_sourse(row):
    if 'yandex' in row['traffic_source'] or 'google' in row['traffic_source']:
        return 'organic'
    if 'email' in row['traffic_source'] or 'paid' in row['traffic_source']:
        if 'Russia' in row['region']:
            return 'ad'
        else:
            return 'other'
    return row['traffic_source']
```

```python
logs['sourse_type'] = logs.apply(type_sourse, axis=1)
```

```python
logs.head(10)
```

```python
logs['sourse_type'].value_counts()
```

```python
rates = pd.read_csv('ml-latest-small/ratings.csv')
```

```python
rates.head()
```

```python
user_rates = rates.pivot_table(index='userId', columns='rating', values='movieId', aggfunc='count', fill_value=0, margins=True)
```

```python
user_rates
```

```python
all_rates = rates.merge(user_rates, on='userId')
```

```python
all_rates = all_rates.loc[all_rates.All >= 100]
```

```python
all_rates[ all_rates.userId == 536].timestamp.max() - all_rates[ all_rates.userId == 536].timestamp.min()
```

```python
def live_time(data):
    return data.timestamp.max() - data.timestamp.min()
```

```python
all_rates.groupby('userId').apply(live_time).head()
```

```python
mean_of_timelive = int(all_rates.groupby('userId').apply(live_time).mean())
```

```python
mean_of_timelive
```
