```python
def data_article(article):
    '''Ищет дату публикации статьи'''
    time_p = article.find('span', class_='post__time').text
    return time_p
```

```python
def header_article(article):
    '''Ищет заголовок статьи '''
    header = article.find('a', class_='post__title_link').text
    return header
```

```python
def link_article(article):
    '''Ищет ссылку на статью'''
    link = article.find('h2', 'post__title')
    link = link.find('a').get('href')
    return link
```

```python
def text_article(link):
    '''Ищет полный текст статьи'''
    text_a1 = requests.get(link)
    text_a2 = BeautifulSoup(text_a1.text)
    text_a3 = text_a2.find('div', class_='post__text').text
    return text_a3
```
