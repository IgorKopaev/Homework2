```python
horses_selection = horses[[1, 2, 4, 5, 6, 7, 11, 23]]
horses_selection = horses_selection.rename(columns={
                            1: 'surgery', 2: 'age', 4: 'rectal temperature', 5: 'pulse',
                            6: 'respiratory rate', 7: 'temperature of extremities', 11: 'pain', 23: 'outcome'})
horses_selection.head()
```

```python
horses_selection.describe()
```

```python
def blowouts(column):
    q1 = horses_selection[column].quantile(0.25)
    q3 = horses_selection[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = horses_selection[horses_selection[column].between(lower_bound, upper_bound, inclusive=True)]
    blowout = pd.concat([horses_selection, remove_outliers]).drop_duplicates(keep=False).dropna()
    return blowout
```

```python
def no_blowouts(column):
    q1 = horses_selection[column].quantile(0.25)
    q3 = horses_selection[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = horses_selection[horses_selection[column].between(lower_bound, upper_bound, inclusive=True)]
    return remove_outliers
```

```python
horses_selection_m['surgery'].fillna(horses_selection_m['surgery'].mode()[0], inplace=True)
horses_selection_m['surgery'].value_counts()
#Так как пропуск в этом стоблце один, то замена его на моду не повлияет на репрезентативность
```

```python
print(horses_selection_m.groupby('temperature of extremities')['rectal temperature'].median())
temp = horses_selection_m.groupby('temperature of extremities')['rectal temperature'].transform('median')
horses_selection_m['rectal temperature'].fillna(temp, inplace=True)
'''Здесь стоит проследить за взаимосвязью температуры конечностей и температуры общей, и поставтиь в пропусках
отгруппированные медианы, чтобы значения соотносились. Средение не сильно подходит из-за большого числа выбросов'''
```

```python
print(horses_selection_m.groupby('temperature of extremities')['pulse'].mean())
pul = horses_selection_m.groupby('temperature of extremities')['pulse'].transform('mean')
horses_selection_m['pulse'].fillna(pul, inplace=True)
'''Аналогично с температурой тела, но здесь выбросов меньше и они соотносятся с краями квантилей, поэтому здесб лучше средение'''
```

```python
print(horses_selection_m.groupby('pain')['respiratory rate'].describe())
resp = horses_selection_m.groupby('pain')['respiratory rate'].transform('median')
horses_selection_m['respiratory rate'].fillna(resp, inplace=True)
'''Много выбросов в конце выборки, это двигает средение арифметическое вправо, лучше медиана будет.
Болевые ощущения могут влиять на частоту дыхания, и можно увидеть, что в случае средней боли или её отстуствии дыхание
становится спокойнее'''
```
